# Tuning dos parâmetros

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score, KFold
import pickle
```

```python
with open('BR-21.pkl', 'rb') as f:
    X, y = pickle.load(f)
```

---

### Decision Tree

```python
params = {'criterion': ['gini', 'entropy'],
                   'splitter': ['best', 'random'],
                   'min_samples_split': [2,5,10],
                  'min_samples_leaf': [1,5,10]}

grid_search = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=params, cv=5)
grid_search.fit(X, y)
tree_best_params = grid_search.best_params_
tree_best_score = grid_search.best_score_
print(tree_best_params)
print(tree_best_score)
```

-----

### Random Forest

```python
params = {'criterion': ['gini', 'entropy'],
                   'n_estimators': [10,30,50,70,100],
                   'min_samples_split': [2,5,10],
                  'min_samples_leaf': [1,5,10]}

grid_search = GridSearchCV(estimator=RandomForestClassifier(), param_grid=params, cv=5)
grid_search.fit(X, y)
rf_best_params = grid_search.best_params_
rf_best_score = grid_search.best_score_
print(rf_best_params)
print(rf_best_score)
```

-----

### Knn

```python
params = {'n_neighbors': [3,5,10,20],
         'p': [1,2,5,10,20]}

grid_search = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=params, cv=5)
grid_search.fit(X, y)
knn_best_params = grid_search.best_params_
knn_best_score = grid_search.best_score_
print(knn_best_params)
print(knn_best_score)
```

----

### Regressão Logística

```python
params = {'tol': [0.0001,0.00001,0.000001],
         'C': [1.0,1.5,2.0],
         'solver': ['lbfgs', 'sag', 'saga'],
         'max_iter': [500]}

grid_search = GridSearchCV(estimator=LogisticRegression(), param_grid=params, cv=5)
grid_search.fit(X, y)
logreg_best_params = grid_search.best_params_
logreg_best_score = grid_search.best_score_
```

```python
print(logreg_best_params)
print(logreg_best_score)
```

---

### SVM

```python
params = {'tol': [0.001,0.0001,0.00001],
         'C': [1.0,1.5,2.0],
         'kernel': ['rbf','linear','poly','sigmoid']}

grid_search = GridSearchCV(estimator=SVC(), param_grid=params, cv=5)
grid_search.fit(X, y)
svm_best_params = grid_search.best_params_
svm_best_score = grid_search.best_score_
print(svm_best_params)
print(svm_best_score)
```

----

### Redes Neurais

```python
params = {'activation': ['relu', 'logistic', 'tahn'],
         'solver': ['adam', 'sgd'],
#          'batch_size': [10,20,30],
         'max_iter': [3000],
         'hidden_layer_sizes': [(2,2),(10,10),(100,100)],
         'tol': [0.0000100]}

grid_search = GridSearchCV(estimator=MLPClassifier(), param_grid=params, cv=5)
grid_search.fit(X, y)
neural_best_params = grid_search.best_params_
neural_best_score = grid_search.best_score_
```

```python
print(neural_best_params)
print(neural_best_score)
```

----

# Cross Validation

Agora, vamos fazer a validação cruzada utilizando o melhor tuning possível, com random_states diferentes.

```python
results_naive = []
results_tree = []
results_random_forest = []
results_knn = []
results_logistic = []
results_svc = []
results_rede_neural = []

for i in range(30): # 30 é um número bem aceito  na comunidade científica
    kfold = KFold(n_splits=5, shuffle=True, random_state=i) # a base vai ser dividida em 10 pedaços
    
    naive = GaussianNB()
    scores = cross_val_score(naive, X, y, cv=kfold)
    results_naive.append(scores.mean())

    tree = DecisionTreeClassifier(criterion=tree_best_params['criterion'], min_samples_leaf=tree_best_params['min_samples_leaf'], min_samples_split=tree_best_params['min_samples_split'], splitter=tree_best_params['splitter'])
    scores = cross_val_score(tree, X, y, cv=kfold)
    # print(scores)
    # print(scores.mean())
    results_tree.append(scores.mean())
    
    random_forest = RandomForestClassifier(criterion=rf_best_params['criterion'], min_samples_leaf=rf_best_params['min_samples_leaf'], min_samples_split=rf_best_params['min_samples_split'], n_estimators=rf_best_params['n_estimators'])
    scores = cross_val_score(random_forest, X, y, cv=kfold)
    results_random_forest.append(scores.mean())
    
    knn = KNeighborsClassifier(n_neighbors=knn_best_params['n_neighbors'], p=knn_best_params['p'])
    scores = cross_val_score(knn, X, y, cv=kfold)
    results_knn.append(scores.mean())
    
    logistic = LogisticRegression(C=logreg_best_params['C'], solver=logreg_best_params['solver'], tol=logreg_best_params['tol'], max_iter=logreg_best_params['max_iter'])
    scores = cross_val_score(logistic, X, y, cv=kfold)
    results_logistic.append(scores.mean())
    
    svc = SVC(C=svm_best_params['C'], kernel=svm_best_params['kernel'], tol=svm_best_params['tol'])
    scores = cross_val_score(svc, X, y, cv=kfold)
    results_svc.append(scores.mean())
    
    rede_neural = MLPClassifier(activation=neural_best_params['activation'], max_iter=neural_best_params['max_iter'], solver=neural_best_params['solver'], hidden_layer_sizes=neural_best_params['hidden_layer_sizes'], tol=neural_best_params['tol'])
    scores = cross_val_score(rede_neural, X, y, cv=kfold)
    results_rede_neural.append(scores.mean())
    print(i)
```

```python
results = pd.DataFrame({'Naive': results_naive, 'Tree': results_tree, 'Random Forest': results_random_forest, 'Knn': results_knn, 'LogReg': results_logistic, 'SVM': results_svc, 'Neural Network': results_rede_neural})
results
```
